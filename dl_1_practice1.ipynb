```python
df = df.drop(['Unnamed: 15','Unnamed: 16'], axis=1)
```

```python
df.drop(columns=['CAT. MEDV'], inplace=True)
```

```python
df
```

```python
fig = plt.figure(figsize=(14,8))
sns.heatmap(df.corr().round(2), annot=True)
```

```python
X = df[['LSTAT', 'RM']]
X
```

```python
Y = df['MEDV']
Y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.25,random_state=50)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
sc.fit(x_train)
```

```python
x_train = sc.transform(x_train)
x_test = sc.transform(x_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(128, input_shape=(2,), activation='relu', name='input'))
model.add(Dense(64, activation='relu', name='layer_1'))
model.add(Dense(1, activation='linear', name='output'))
model.compile(optimizer='adam', loss='mse', metrics='mae')
model.summary()
```

```python
model.fit(x_train, y_train, epochs=100, validation_split=0.05)
```

```python
output = model.evaluate(x_test, y_test)
```

```python
print("MSE = ",output[0])
print("MAE = ",output[1])
```

```python
y_pred = model.predict(x=x_test)
```
